## Criando uma tabela fato da distribuição de medicamentos


```python
import pandas as pd
import numpy as np
import random

# --- CONFIGURAÇÕES GERAIS ---
random.seed(42)
np.random.seed(42)

# Definição de Capitais para Regra de Negócio
CAPITAIS_NE = [
    "Recife", "Salvador", "Fortaleza", "São Luís", "Maceió",
    "João Pessoa", "Teresina", "Natal", "Aracaju"
]

# --- 1. CARREGAMENTO DAS TABELAS DIMENSÃO ---

try:
    # 1. Carregar Farmácias (108 lojas)
    #
    df_farmacias = pd.read_csv("dim_farmacias_final.csv")

    # 2. Carregar Lista Mestra de Medicamentos (3080 SKUs)
    # Ajuste o caminho conforme necessário
    df_produtos = pd.read_csv("lista_medicamentos_expandida.csv")

    print(f"Farmácias carregadas: {len(df_farmacias)}")
    print(f"Produtos carregados: {len(df_produtos)}")

except FileNotFoundError as e:
    print(f"ERRO CRÍTICO: Arquivo não encontrado - {e}")
    # Mock para não quebrar o código se você for testar sem arquivos
    df_farmacias = pd.DataFrame([{'cod_farmacia': 1, 'nome_farmacia': 'Teste', 'cidade': 'Recife'}])
    df_produtos = pd.DataFrame([{'cod_medicamento': 700000, 'nome_medicamento': 'Dipirona', 'categoria': 'Isentos', 'fabricante': 'EMS', 'dosagem': '500mg', 'apresentacao': 'Cp', 'substancia_principal': 'Dipirona'}])

# --- 2. GERAÇÃO DO DATASET DE ESTOQUE ---

print("Gerando estoque distribuído (pode levar alguns segundos)...")

dados_estoque = []

# garantir cobertura
for _, farmacia in df_farmacias.iterrows():

    eh_capital = farmacia['cidade'] in CAPITAIS_NE

    for _, produto in df_produtos.iterrows():

        categoria = produto['categoria']

        # --- LÓGICA DE QUANTIDADE BASE ---
        # Produtos de alta saída tem base maior
        if categoria in ["Higiene e Beleza", "Medicamentos Isentos", "Diversos"]:
            qtd_base = random.randint(20, 150)
        elif categoria == "Uso Contínuo":
            qtd_base = random.randint(15, 80)
        elif categoria == "Antibióticos e Controlados":
            qtd_base = random.randint(5, 40)
        elif categoria == "Refrigerados": # Vacinas/Insulinas
            qtd_base = random.randint(2, 15)
        else:
            qtd_base = random.randint(5, 20)

        # --- MULTIPLICADOR DE LOCALIZAÇÃO ---
        if eh_capital:
            # Capitais tem estoque 30% a 80% maior
            fator_loc = random.uniform(1.3, 1.8)
        else:
            # Interior tem estoque 20% a 50% menor
            fator_loc = random.uniform(0.5, 0.8)

        # --- REGRA DE RARIDADE / ESTOQUE ZERO ---
        # Medicamentos raros/caros podem não ter no interior
        qtd_final = int(qtd_base * fator_loc)

        if not eh_capital:
            # Se for interior, chance de não ter estoque de itens complexos
            if categoria in ["Refrigerados", "Antibióticos e Controlados"]:
                if random.random() < 0.15: # 15% de chance de ser zero
                    qtd_final = 0
            # Mesmo itens comuns podem faltar ocasionalmente
            elif random.random() < 0.02:
                qtd_final = 0

        # Montagem da linha
        row = {
            # Dados do Produto
            "cod_medicamento": produto["cod_medicamento"],
            "nome_medicamento": produto["nome_medicamento"],
            "substancia_principal": produto["substancia_principal"],
            "apresentacao": produto["apresentacao"],
            "dosagem": produto["dosagem"],
            "fabricante": produto["fabricante"],
            "categoria": produto["categoria"],

            # Dados de Estoque
            "quantidade_estoque": qtd_final,

            # Dados da Farmácia
            "cod_farmacia": farmacia["cod_farmacia"],
            "nome_farmacia": farmacia["nome_farmacia"],
            "cidade": farmacia["cidade"]
        }
        dados_estoque.append(row)

# Criar DataFrame
df = pd.DataFrame(dados_estoque)

# --- 3. INTRODUÇÃO DE IMPERFEIÇÕES ---

def introduzir_erro(valor, chance=0.03, tipo='texto'):
    if pd.isna(valor): return valor
    prob = random.random()
    if prob < chance:
        if tipo == 'nan': return np.nan
        if tipo == 'texto_upper': return str(valor).upper()
        if tipo == 'texto_lower': return str(valor).lower()
    return valor

# Aplicar imperfeições solicitadas
# A. Dados Faltantes (~2%)
cols_sujas_nan = ['nome_farmacia', 'substancia_principal', 'fabricante', 'apresentacao']
for col in cols_sujas_nan:
    if col in df.columns:
        df[col] = df[col].apply(lambda x: introduzir_erro(x, chance=0.02, tipo='nan'))

# B. Sujeira na Cidade (~1% nulo, ~15% formatação)
df['cidade'] = df['cidade'].apply(lambda x: introduzir_erro(x, chance=0.01, tipo='nan'))
df['cidade'] = df['cidade'].apply(lambda x: introduzir_erro(x, chance=0.10, tipo='texto_upper'))
df['cidade'] = df['cidade'].apply(lambda x: introduzir_erro(x, chance=0.05, tipo='texto_lower'))

# C. Erros na Quantidade
def sujar_quantidade(val):
    prob = random.random()
    if prob < 0.001: return -val      # Valor negativo (raro)
    if prob < 0.001: return 99999     # Outlier (raro)
    return val

df['quantidade_estoque'] = df['quantidade_estoque'].apply(sujar_quantidade)

# --- 4. EXPORTAÇÃO ---
print(f"Dataset Final Gerado com {len(df)} registros.")
print(df[['nome_medicamento', 'cidade', 'quantidade_estoque']].head(10))

filename = "distribuicao_medicamentos_completa.csv"
df.to_csv(filename, index=False, encoding='utf-8')
print(f"Arquivo '{filename}' pronto para download.")
```

```python
import pandas as pd
import numpy as np
import random
import calendar
from datetime import datetime, date

# --- 1. CARREGAMENTO E PREPARAÇÃO ---

try:
    distrib = pd.read_csv("distribuicao_medicamentos_completa.csv")
    print(f"Base carregada: {len(distrib)} linhas.")
except FileNotFoundError:
    print("ERRO: Arquivo não encontrado. Verifique o caminho.")
    # Seção de emergência para o código não quebrar em caso de testar sem o arquivo
    distrib = pd.DataFrame({
        'cod_farmacia': [1001]*10, 'cod_medicamento': range(700000, 700010),
        'quantidade_estoque': [50]*10, 'fabricante': ['EMS']*10,
        'cidade': ['Recife']*10, 'nome_farmacia': ['Teste']*10
    })

df_remessas_base = distrib[['cod_farmacia', 'cod_medicamento', 'quantidade_estoque', 'fabricante']].copy()

# --- 2. LÓGICA DE QUANTIDADE ---

# Limpeza para cálculo
def get_clean_stock(val):
    try:
        v = float(val)
        return max(0, int(v))
    except:
        return 0

stock_clean = df_remessas_base['quantidade_estoque'].apply(get_clean_stock)
fator_aumento = np.random.uniform(1.20, 1.50, size=len(stock_clean))
total_semestre = stock_clean * fator_aumento

# Se o estoque for 0, definimos um envio mínimo padrão
total_semestre = np.where(total_semestre == 0, 60, total_semestre)

# Quantidade Fixa Mensal
df_remessas_base['qtd_mensal_fixa'] = np.ceil(total_semestre / 6).astype(int)

# --- 3. LÓGICA DE LABORATÓRIOS E LOTES ---

# A. Lista Oficial de Fabricantes)
laboratorios_oficiais = [
    "NewQuímica", "Medlove", "ENS", "Asiafarma",
    "Tenuto", "Pratik-Donaducci", "Axex", "Sandozan"
]

# Mapa Fixo: Nome -> ID
mapa_lab_id = {nome: 500 + i for i, nome in enumerate(laboratorios_oficiais)}

# Função para corrigir fabricantes nulos/sujos na base original antes de gerar ID
def corrigir_fabricante(val):
    # Se o valor não estiver na lista oficial (é nulo, sujo ou vazio), atribui um aleatório
    if val not in laboratorios_oficiais:
        return random.choice(laboratorios_oficiais)
    return val

# Cria uma coluna temporária limpa para garantir que todo mundo tenha ID
df_remessas_base['fabricante_corrigido'] = df_remessas_base['fabricante'].apply(corrigir_fabricante)

# Gera o ID baseando-se apenas nos nomes corrigidos
df_remessas_base['id_fabricante'] = df_remessas_base['fabricante_corrigido'].map(mapa_lab_id)

# Atualiza a coluna fabricante original para a corrigida
df_remessas_base['fabricante'] = df_remessas_base['fabricante_corrigido']

# B. Gerar Matriz de Lotes (Laboratório + Mês = Lote Único)
lotes_db = {}
meses = [7, 8, 9, 10, 11, 12]
ano = 2025

for lab_id in mapa_lab_id.values():
    for mes in meses:
        letras = "".join(random.choices("ABCDEF", k=2))
        nums = "".join(random.choices("0123456789", k=4))
        # Ex: EMS em Outubro sempre terá lote AB1234-10
        lotes_db[(lab_id, mes)] = f"{letras}{nums}-{mes}"

# --- 4. EXPANSÃO TEMPORAL ---

dfs_mensais = []

for mes in meses:
    # Copia a base calculada
    df_mes = df_remessas_base.copy()

    # 1. Define a Quantidade (é fixa mensal)
    df_mes['quantidade_recebida'] = df_mes['qtd_mensal_fixa']

    # 2. Define a Data de entrega do lote
    ultimo_dia = calendar.monthrange(ano, mes)[1]
    inicio_janela = ultimo_dia - 3

    # Função para gerar dia aleatório na janela de tempo que foi esoecificada
    def gerar_dia_entrega(row):
        dia = random.randint(inicio_janela, ultimo_dia)
        return f"{ano}-{mes:02d}-{dia:02d}"

    df_mes['data_entrega'] = df_mes.apply(gerar_dia_entrega, axis=1)

    # 3. Atribui o Lote
    df_mes['lote_fabricacao'] = df_mes['id_fabricante'].apply(lambda x: lotes_db.get((x, mes), "LOTE-GEN"))

    dfs_mensais.append(df_mes)

# Junta tudo
df_final_remessas = pd.concat(dfs_mensais, ignore_index=True)

# --- 5. AJUSTES FINAIS ---

# ID da Transação Único
df_final_remessas['id_transacao_remessa'] = range(1000000, 1000000 + len(df_final_remessas))

# Limpar colunas auxiliares que não vão para o CSV final
# Adicionado 'fabricante' na lista de colunas a manter
cols_to_keep = [
    'id_transacao_remessa', 'cod_farmacia', 'cod_medicamento',
    'fabricante', 'id_fabricante', # <--- Colunas Solicitadas
    'quantidade_recebida', 'data_entrega', 'lote_fabricacao'
]
df_final_remessas = df_final_remessas[cols_to_keep]

# --- 6. IMPERFEIÇÕES ---

# A. Datas sujas (Vazias ou Futuras Erradas)
def sujar_data(val):
    prob = random.random()
    if prob < 0.01: return "" # Vazio
    if prob < 0.01: return "2030-01-01" # Data errada
    return val

df_final_remessas['data_entrega'] = df_final_remessas['data_entrega'].apply(sujar_data)


# --- 7. EXPORTAÇÃO ---

print(f"Histórico de Remessas (6 meses) gerado com {len(df_final_remessas)} linhas.")
print(df_final_remessas.head(10))

filename = "historico_remessas_2sem.csv"
df_final_remessas.to_csv(filename, index=False, encoding='utf-8')
print(f"Arquivo '{filename}' salvo com sucesso!")
```

## Tabela de vendas e tabela de clientes




```python
import pandas as pd
import numpy as np
import random
import uuid
from datetime import datetime, timedelta

# --- CONFIGURAÇÕES ---
random.seed(42)
np.random.seed(42)

# Datas (2º Semestre 2025)
START_DATE = datetime(2025, 7, 1)
END_DATE = datetime(2025, 12, 31)
DAYS_RANGE = (END_DATE - START_DATE).days

# --- 1. CARREGAR DADOS ---
print("Carregando bases de dados...")
try:
    df_estoque = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/tabelas_pharma/distribuicao_medicamentos_completa.csv")
    df_remessas = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/tabelas_pharma/historico_remessas_2sem.csv")
    df_produtos = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/tabelas_pharma/lista_medicamentos_expandida.csv")

    # Mapeamento de Preço e Categoria para consulta rápida
    prod_info = df_produtos.set_index('cod_medicamento')[['preco_unitario', 'categoria', 'substancia_principal']].to_dict('index')

except FileNotFoundError:
    print("ERRO: Arquivos anteriores não encontrados. Execute os passos anteriores.")
    # Mock para evitar quebra do script se rodado isoladamente
    exit()

# --- 2. CÁLCULO DE VENDAS NECESSÁRIAS (CONSISTÊNCIA MATEMÁTICA) ---
print("Calculando volume de vendas para garantir consistência...")

# A. Total Recebido por Farmácia/Produto
remessas_agg = df_remessas.groupby(['cod_farmacia', 'cod_medicamento'])['quantidade_recebida'].sum().reset_index()
remessas_agg.rename(columns={'quantidade_recebida': 'qtd_recebida_total'}, inplace=True)

# B. Estoque Atual
estoque_agg = df_estoque[['cod_farmacia', 'cod_medicamento', 'quantidade_estoque']].copy()

# C. Cruzamento (Merge)
df_calc = pd.merge(remessas_agg, estoque_agg, on=['cod_farmacia', 'cod_medicamento'], how='left')
df_calc['quantidade_estoque'] = df_calc['quantidade_estoque'].fillna(0)

# D. Lógica: O que entrou - O que sobrou = O que foi vendido
df_calc['qtd_vendida_total'] = df_calc['qtd_recebida_total'] - df_calc['quantidade_estoque']

# Ajuste de segurança (não pode haver vendas negativas, se houver erro no estoque, zeramos a venda)
df_calc['qtd_vendida_total'] = df_calc['qtd_vendida_total'].apply(lambda x: x if x > 0 else 0)

# Filtrar apenas o que tem venda para processar
items_to_sell = df_calc[df_calc['qtd_vendida_total'] > 0].copy()

print(f"Total de itens unitários a serem vendidos na simulação: {items_to_sell['qtd_vendida_total'].sum()}")

# --- 3. GERAÇÃO DA TABELA CLIENTES ---
print("Gerando base de clientes...")

NUM_CLIENTES_CADASTRADOS = 3000
clientes_data = []

nomes_base = ["Maria", "João", "Ana", "Pedro", "Lucas", "Mariana", "Carlos", "Fernanda", "Paulo", "Juliana", "Marcos", "Larissa", "Gabriel", "Camila"]
sobrenomes = ["Silva", "Santos", "Oliveira", "Souza", "Rodrigues", "Ferreira", "Alves", "Pereira", "Lima", "Gomes", "Costa", "Ribeiro"]
dominios = ["gmail.com", "outlook.com", "uol.com.br", "hotmail.com", "bol.com.br"]

def gerar_cpf():
    return "".join([str(random.randint(0, 9)) for _ in range(11)])

for i in range(NUM_CLIENTES_CADASTRADOS):
    nome = f"{random.choice(nomes_base)} {random.choice(sobrenomes)}"
    cpf = gerar_cpf()
    email = f"{nome.lower().replace(' ', '.').replace('ç', 'c')}@{random.choice(dominios)}"

    clientes_data.append({
        "id_cliente": 10000 + i, # IDs 10000 a 12999 são cadastrados
        "nome_cliente": nome,
        "cpf": cpf,
        "email": email,
        "perfil_comprador": random.choice(["Uso Contínuo", "Geral", "Geral"])
    })

df_clientes = pd.DataFrame(clientes_data)

# Lista de IDs para sorteio
ids_cadastrados = df_clientes['id_cliente'].tolist()

# --- 4. MOTOR DE SIMULAÇÃO DE VENDAS (POR FARMÁCIA) ---
print("Simulando transações (isso pode levar um minuto)...")

transacoes_list = []
transacao_id_counter = 5000000
cliente_anonimo_counter = 9000000 # IDs altos para não identificados

# Agrupar itens por farmácia para processar loja a loja
grouped_lojas = items_to_sell.groupby('cod_farmacia')

metodos_pagamento = ["Cartão de Crédito", "Cartão de Débito", "Dinheiro", "PIX", "Farmácia Popular"]

for cod_farmacia, group in grouped_lojas:

    # Cria uma "Pilha de Produtos" para esta farmácia
    # Ex: se temos 5 caixas de Dipirona para vender, adicionamos o ID da dipirona 5 vezes na lista
    # Isso permite sortear produtos da prateleira aleatoriamente para compor cestas
    pilha_produtos = []

    for _, row in group.iterrows():
        cod_med = row['cod_medicamento']
        qtd = int(row['qtd_vendida_total'])
        pilha_produtos.extend([cod_med] * qtd)

    # Embaralha a pilha para vendas aleatórias
    random.shuffle(pilha_produtos)

    # Processa a pilha criando cestas (transações)
    while len(pilha_produtos) > 0:

        # 1. Definir Tamanho da Cesta (1 a 7 itens, com peso maior em cestas pequenas)
        basket_size = random.choices(
            [1, 2, 3, 4, 5, 6, 7, 10, 15],
            weights=[30, 25, 15, 10, 8, 5, 4, 2, 1]
        )[0]

        # Ajusta se sobrar poucos itens
        if basket_size > len(pilha_produtos):
            basket_size = len(pilha_produtos)

        # 2. Retirar itens da pilha
        cesta = [pilha_produtos.pop() for _ in range(basket_size)]

        # 3. Validar Regras da Cesta (Controlados não podem ter qtd absurda)
        # Se a cesta tiver muitos controlados iguais, espalha para próxima venda (simplificação: aceitamos por agora pois a pilha foi embaralhada)

        # 4. Definir Cliente
        is_identificado = random.random() > 0.55 # 55% de chance de ser não identificado

        if is_identificado:
            id_cliente = random.choice(ids_cadastrados)
        else:
            id_cliente = cliente_anonimo_counter
            cliente_anonimo_counter += 1

        # 5. Definir Data
        days_offset = random.randint(0, DAYS_RANGE)
        data_venda = START_DATE + timedelta(days=days_offset)

        # 6. Calcular Totais
        valor_total = 0
        codigos_na_cesta = []

        # Agrupa itens iguais na mesma transação (ex: comprou 2 dipironas)
        contagem_cesta = {}
        for item in cesta:
            contagem_cesta[item] = contagem_cesta.get(item, 0) + 1

        for cod_med, qtd in contagem_cesta.items():
            # Busca preço e multiplica
            info = prod_info.get(cod_med, {'preco_unitario': 10.0}) # Fallback safe
            valor_total += info['preco_unitario'] * qtd

            # Formata string de códigos (repetindo o código se comprou mais de 1, ou agrupado? O prompt pede "pode conter mais de um código")
            # Vamos listar todos os códigos: "7001, 7001, 7005"
            codigos_na_cesta.extend([str(cod_med)] * qtd)

        # Formatação final da linha
        transacoes_list.append({
            "id_transacao": transacao_id_counter,
            "cod_farmacia": cod_farmacia,
            "cod_medicamento": ",".join(codigos_na_cesta), # Lista como string
            "qtd_produtos": len(cesta),
            "data_venda": data_venda.strftime("%Y-%m-%d"),
            "valor_total_compra": round(valor_total, 2),
            "metodo_pagamento": random.choice(metodos_pagamento),
            "id_cliente": id_cliente
        })

        transacao_id_counter += 1

# --- 5. EXPORTAÇÃO ---

# Tabela Vendas
df_vendas = pd.DataFrame(transacoes_list)
print(f"Vendas geradas: {len(df_vendas)} transações.")
print("Exemplo de Venda:")
print(df_vendas.head())

filename_vendas = "tabela_vendas.csv"
df_vendas.to_csv(filename_vendas, index=False, encoding='utf-8')

# Tabela Clientes (Apenas os identificados para a dimensão)
# Os anonimos (ID > 9000000) não vão para a tabela dimensão, simulando que não tem cadastro
filename_clientes = "dim_clientes.csv"
# Removemos a coluna auxiliar de perfil antes de salvar
df_clientes.drop(columns=['perfil_comprador']).to_csv(filename_clientes, index=False, encoding='utf-8')

print(f"Arquivos '{filename_vendas}' e '{filename_clientes}' salvos com sucesso!")
```
